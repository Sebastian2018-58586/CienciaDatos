## DistribuciÃ³n de la Memoria RAM de Laptops

#### Analisis
El mercado de laptops estÃ¡ dominado por modelos de gama media. Sin embargo, las laptops de gama alta siguen representando un mercado de nicho, con un pequeÃ±o porcentaje de La mayorÃ­a de las laptops probablemente estÃ©n equipadas con entre 4 GB y 16 GB de RAM, que es la cantidad mÃ¡s comÃºnmente utilizada para tareas generales y juegos ligeros.

#### Distribuciones de las Variables NumÃ©ricas de Laptops

```python

# Lista de 7 variables numÃ©ricas que deseas visualizar
numeric_vars = ['Price_euros', 'Weight', 'ScreenW', 'ScreenH', 'CPU_freq', 'PrimaryStorage', 'SecondaryStorage']

# Reemplazar ceros por NaN y luego por la mediana (solo para ScreenW y ScreenH si es necesario)
for col in ['ScreenW', 'ScreenH']:
    laptop[col] = laptop[col].replace(0, np.nan)
    laptop[col].fillna(laptop[col].median(), inplace=True)

# Crear grid de 2x4 = 8, pero usar solo 7 ejes
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

# Graficar los histogramas
for i, col in enumerate(numeric_vars):
    sns.histplot(data=laptop, x=col, bins=30, kde=True, ax=axes[i], color='royalblue')
    axes[i].set_title(f'DistribuciÃ³n de {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frecuencia')

# Eliminar el Ãºltimo subplot vacÃ­o si es necesario
for j in range(len(numeric_vars), len(axes)):
    fig.delaxes(axes[j])

# TÃ­tulo general
plt.suptitle('Distribuciones de Variables NumÃ©ricas Seleccionadas en Laptops', fontsize=16, y=1.03)
plt.tight_layout()
plt.show()

```

#### AnÃ¡lisis de Outliers:
Price_euros:
El precio de las laptops se concentra entre 300â‚¬ y 1500â‚¬, aunque se observan outliers notables por encima de 2000â‚¬, correspondientes a equipos de gama alta.

Weight:
La mayorÃ­a de laptops pesan entre 1 y 2.5 kg, pero existen outliers ligeros (< 1kg) y pesados (> 3.5 kg) que podrÃ­an corresponder a ultraligeros o estaciones de trabajo mÃ³viles.

ScreenW / ScreenH:
Las resoluciones estÃ¡ndar estÃ¡n bien representadas, pero existen algunos valores extremos de resoluciÃ³n que pueden indicar configuraciones premium o errores.

CPU_freq:
Aunque la frecuencia promedio ronda los 2.2 - 2.5 GHz, hay valores inusualmente altos (> 4 GHz) que podrÃ­an ser errores o CPUs con overclocking.

PrimaryStorage / SecondaryStorage:
Se identifican outliers marcados con configuraciones de almacenamiento muy altas, particularmente por encima de 1TB, que reflejan laptops para gaming o ediciÃ³n de video.

```python

# Variables numÃ©ricas clave del dataset de laptops
numeric_vars = ['Price_euros', 'Weight', 'ScreenW', 'ScreenH', 'CPU_freq', 'PrimaryStorage', 'SecondaryStorage']

# Limpiar ceros que no tienen sentido en dimensiones de pantalla
laptop['ScreenW'] = laptop['ScreenW'].replace(0, np.nan)
laptop['ScreenH'] = laptop['ScreenH'].replace(0, np.nan)

# Crear grid de boxplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.ravel()

for i, col in enumerate(numeric_vars):
    sns.boxplot(data=laptop, y=col, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot de {col}', fontsize=12)
    axes[i].set_ylabel('')
    axes[i].grid(True)

# Eliminar grÃ¡fico vacÃ­o (hay 7 variables y 8 subplots)
fig.delaxes(axes[-1])

plt.suptitle('Boxplots de Variables NumÃ©ricas Clave - Laptop Prices', fontsize=16, y=1.03)
plt.tight_layout()
plt.show()
```

Las "Ultrabook" tambiÃ©n presentan un precio elevado, debido a su diseÃ±o ligero y componentes premium.

Tendencia por aÃ±o:

En general, los precios mÃ¡s altos se concentran en los Ãºltimos aÃ±os registrados (por ejemplo, 2017 o 2018 si estÃ¡n presentes en tu dataset).
Tipos como "2 in 1 Convertible" y "Ultrabook" han mantenido una alta variabilidad en precios, reflejando su diversidad de configuraciones.

Estabilidad en ciertos tipos:

## 2. AnÃ¡lisis de Correlaciones con Heatmaps

Correlaciones Destacadas:

ScreenW â†” ScreenH: 0.99
RelaciÃ³n prÃ¡cticamente perfecta, lo que indica una fuerte dependencia entre el ancho y el alto de pantalla (resoluciÃ³n).

Price_euros â†” Ram: â‰ˆ 0.70
Muestra una correlaciÃ³n positiva considerable: a mayor memoria RAM, tiende a aumentar el precio.

Price_euros â†” CPU_freq: â‰ˆ 0.45
TambiÃ©n se observa una correlaciÃ³n moderada: procesadores con mayor frecuencia tienden a encontrarse en laptops mÃ¡s costosas.

Peso vs TamaÃ±o de pantalla (Inches):
Existe una relaciÃ³n positiva moderada, lo que sugiere que laptops mÃ¡s grandes tambiÃ©n suelen ser mÃ¡s pesadas.

Almacenamientos (Primary/Secondary):
Tienen correlaciones dÃ©biles con el precio, indicando que no son el factor determinante principal.

```python

# SelecciÃ³n de variables numÃ©ricas relevantes
numeric_cols = ['Inches', 'Ram', 'Weight', 'Price_euros', 'ScreenW', 'ScreenH', 'CPU_freq', 'PrimaryStorage', 'SecondaryStorage']

# Matriz de correlaciÃ³n
corr_matrix = laptop[numeric_cols].corr()

# Heatmap de correlaciÃ³n
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Matriz de CorrelaciÃ³n entre Variables NumÃ©ricas', fontsize=16)
plt.tight_layout()
plt.show()
```

#### CorrelaciÃ³n Precio vs Variables Clave (Scatter Plots)
Hallazgos Principales:

Ram vs Precio:
Se aprecia una relaciÃ³n positiva clara: a mayor cantidad de RAM, el precio tiende a aumentar.
Las laptops con 16GB o mÃ¡s estÃ¡n claramente en rangos de precios mÃ¡s altos.

TamaÃ±o de Pantalla (Inches):
La relaciÃ³n es moderada: laptops mÃ¡s grandes suelen costar mÃ¡s, aunque hay muchas excepciones con precios altos en tamaÃ±os medianos (13-15").

Peso (Weight):
Tendencia ligeramente positiva, pero no muy fuerte.
Laptops mÃ¡s pesadas tienden a tener mayor precio, lo cual puede reflejar mejores materiales o hardware mÃ¡s robusto.

Frecuencia del CPU (CPU_freq):
RelaciÃ³n positiva visible: procesadores de mayor frecuencia suelen encontrarse en laptops mÃ¡s caras.

ResoluciÃ³n de Pantalla (ScreenW / ScreenH):
TambiÃ©n muestran relaciones positivas: resoluciones mÃ¡s altas (Full HD, Retina, etc.) suelen estar asociadas a equipos premium.

```python

# Variables a analizar respecto al precio
features = ['Ram', 'Inches', 'Weight', 'CPU_freq', 'ScreenW', 'ScreenH']

# Crear grÃ¡fico grid
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

for i, feature in enumerate(features):
    sns.scatterplot(data=laptop, x=feature, y='Price_euros', ax=axes[i], color='mediumslateblue')
    axes[i].set_title(f'Precio vs {feature}', fontsize=12)

plt.suptitle('RelaciÃ³n entre Precio y Variables NumÃ©ricas', fontsize=16)
plt.tight_layout()
plt.show()
```
