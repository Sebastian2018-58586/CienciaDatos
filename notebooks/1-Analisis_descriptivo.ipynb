# Histograma de variables numericas

## Distribución de la Memoria RAM de Laptops

#### Analisis
El mercado de laptops está dominado por modelos de gama media. Sin embargo, las laptops de gama alta siguen representando un mercado de nicho, con un pequeño porcentaje de La mayoría de las laptops probablemente estén equipadas con entre 4 GB y 16 GB de RAM, que es la cantidad más comúnmente utilizada para tareas generales y juegos ligeros.
Las laptops con 32 GB o más de RAM serán menos comunes, probablemente encontrándose en modelos de alto rendimiento (estaciones de trabajo, laptops para edición de video, etc.).Si la distribución es sesgada hacia la derecha, podría indicar que las laptops con menos RAM son más comunes.
#### Conclusion
 La gama media de laptops probablemente tiene entre 4 GB y 16 GB de RAM, lo que es suficiente para tareas cotidianas. Los modelos con más de 32 GB de RAM serán de nicho y más caros.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=laptop, x='Ram', bins=20, kde=True, color='blue')
plt.title('Distribución de la Memoria RAM de Laptops', fontsize=14)
plt.xlabel('Memoria RAM (GB)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```

### Histograma de Pantalla (Inches)

Los tamaños más pequeños (por debajo de 13 pulgadas) generalmente están presentes en ultrabooks o laptops ultraportátiles.

###  Distribución de Precios de Laptops (Escala Logarítmica)

#### Distribuciones de las Variables Numéricas de Laptops

```python

# Lista de 7 variables numéricas que deseas visualizar
numeric_vars = ['Price_euros', 'Weight', 'ScreenW', 'ScreenH', 'CPU_freq', 'PrimaryStorage', 'SecondaryStorage']

# Reemplazar ceros por NaN y luego por la mediana (solo para ScreenW y ScreenH si es necesario)
for col in ['ScreenW', 'ScreenH']:
    laptop[col] = laptop[col].replace(0, np.nan)
    laptop[col].fillna(laptop[col].median(), inplace=True)

# Crear grid de 2x4 = 8, pero usar solo 7 ejes
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

# Graficar los histogramas
for i, col in enumerate(numeric_vars):
    sns.histplot(data=laptop, x=col, bins=30, kde=True, ax=axes[i], color='royalblue')
    axes[i].set_title(f'Distribución de {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frecuencia')

# Eliminar el último subplot vacío si es necesario
for j in range(len(numeric_vars), len(axes)):
    fig.delaxes(axes[j])

# Título general
plt.suptitle('Distribuciones de Variables Numéricas Seleccionadas en Laptops', fontsize=16, y=1.03)
plt.tight_layout()
plt.show()

```

#### Análisis de Outliers:
Price_euros:
El precio de las laptops se concentra entre 300€ y 1500€, aunque se observan outliers notables por encima de 2000€, correspondientes a equipos de gama alta.

Weight:
La mayoría de laptops pesan entre 1 y 2.5 kg, pero existen outliers ligeros (< 1kg) y pesados (> 3.5 kg) que podrían corresponder a ultraligeros o estaciones de trabajo móviles.

ScreenW / ScreenH:
Las resoluciones estándar están bien representadas, pero existen algunos valores extremos de resolución que pueden indicar configuraciones premium o errores.

CPU_freq:
Aunque la frecuencia promedio ronda los 2.2 - 2.5 GHz, hay valores inusualmente altos (> 4 GHz) que podrían ser errores o CPUs con overclocking.

PrimaryStorage / SecondaryStorage:
Se identifican outliers marcados con configuraciones de almacenamiento muy altas, particularmente por encima de 1TB, que reflejan laptops para gaming o edición de video.

```python

# Variables numéricas clave del dataset de laptops
numeric_vars = ['Price_euros', 'Weight', 'ScreenW', 'ScreenH', 'CPU_freq', 'PrimaryStorage', 'SecondaryStorage']

# Limpiar ceros que no tienen sentido en dimensiones de pantalla
laptop['ScreenW'] = laptop['ScreenW'].replace(0, np.nan)
laptop['ScreenH'] = laptop['ScreenH'].replace(0, np.nan)

# Crear grid de boxplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.ravel()

for i, col in enumerate(numeric_vars):
    sns.boxplot(data=laptop, y=col, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot de {col}', fontsize=12)
    axes[i].set_ylabel('')
    axes[i].grid(True)

# Eliminar gráfico vacío (hay 7 variables y 8 subplots)
fig.delaxes(axes[-1])

plt.suptitle('Boxplots de Variables Numéricas Clave - Laptop Prices', fontsize=16, y=1.03)
plt.tight_layout()
plt.show()
```

#### Hallazgos Clave:

Diferencia de precios por tipo:

Las laptops "Gaming" y "Workstation" tienen precios significativamente más altos que otras categorías como "Notebook" o "Netbook".
Las "Ultrabook" también presentan un precio elevado, debido a su diseño ligero y componentes premium.

Tendencia por año:

En general, los precios más altos se concentran en los últimos años registrados (por ejemplo, 2017 o 2018 si están presentes en tu dataset).
Tipos como "2 in 1 Convertible" y "Ultrabook" han mantenido una alta variabilidad en precios, reflejando su diversidad de configuraciones.

Estabilidad en ciertos tipos:
Tipos como "Netbook" presentan precios más estables y bajos, con menos outliers y menor dispersión.

```python
# Crear una nueva columna que refleje el tipo de laptop según las columnas dummy
laptop['Laptop_Type'] = laptop[['TypeName_Gaming', 'TypeName_Netbook', 'TypeName_Notebook', 
                                 'TypeName_Ultrabook', 'TypeName_Workstation']].idxmax(axis=1)

# Ahora graficamos el precio según esta nueva columna
plt.figure(figsize=(14, 7))
sns.boxplot(data=laptop, x='Laptop_Type', y='Price_euros', palette='Set2')
plt.title('Boxplot de Precio de Laptops por Tipo', fontsize=16)
plt.xlabel('Tipo de Laptop', fontsize=12)
plt.ylabel('Precio (euros)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()


```

#### Outliers por Marca
Insights por Fabricante:

Precios elevados:
Marcas como Apple y Razer presentan precios significativamente más altos en comparación con otras marcas, destacándose por sus gamas premium.

Variabilidad moderada:
Fabricantes como HP, Lenovo y Dell muestran una amplia gama de precios, lo que sugiere que ofrecen tanto modelos básicos como de alto rendimiento.

Outliers extremos:
Algunas marcas tienen puntos de precio excepcionalmente altos, que podrían corresponder a modelos gamer o con características muy especializadas (por ejemplo, tarjetas gráficas dedicadas, pantallas 4K, etc.).

```python
# Si las columnas de la marca ya son dummies, necesitas aplicar un proceso similar para obtener la marca correspondiente
laptop['Brand'] = laptop[['Company_Apple', 'Company_Asus', 'Company_Dell', 'Company_HP', 
                          'Company_Lenovo', 'Company_MSI', 'Company_Mediacom', 
                          'Company_Samsung', 'Company_Toshiba']].idxmax(axis=1)

# Graficar
plt.figure(figsize=(14, 7))
sns.boxplot(data=laptop, x='Brand', y='Price_euros', palette='pastel')
plt.title('Distribución de Precios de Laptops por Marca (Outliers por Fabricante)', fontsize=16)
plt.xlabel('Marca de Laptop', fontsize=12)
plt.ylabel('Precio (euros)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

## 2. Análisis de Correlaciones con Heatmaps

Correlaciones Destacadas:

ScreenW ↔ ScreenH: 0.99
Relación prácticamente perfecta, lo que indica una fuerte dependencia entre el ancho y el alto de pantalla (resolución).

Price_euros ↔ Ram: ≈ 0.70
Muestra una correlación positiva considerable: a mayor memoria RAM, tiende a aumentar el precio.

Price_euros ↔ CPU_freq: ≈ 0.45
También se observa una correlación moderada: procesadores con mayor frecuencia tienden a encontrarse en laptops más costosas.

Peso vs Tamaño de pantalla (Inches):
Existe una relación positiva moderada, lo que sugiere que laptops más grandes también suelen ser más pesadas.

Almacenamientos (Primary/Secondary):
Tienen correlaciones débiles con el precio, indicando que no son el factor determinante principal.

```python

# Selección de variables numéricas relevantes
numeric_cols = ['Inches', 'Ram', 'Weight', 'Price_euros', 'ScreenW', 'ScreenH', 'CPU_freq', 'PrimaryStorage', 'SecondaryStorage']

# Matriz de correlación
corr_matrix = laptop[numeric_cols].corr()

# Heatmap de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Matriz de Correlación entre Variables Numéricas', fontsize=16)
plt.tight_layout()
plt.show()
```

#### Correlación Precio vs Variables Clave (Scatter Plots)
Hallazgos Principales:

Ram vs Precio:
Se aprecia una relación positiva clara: a mayor cantidad de RAM, el precio tiende a aumentar.
Las laptops con 16GB o más están claramente en rangos de precios más altos.

Tamaño de Pantalla (Inches):
La relación es moderada: laptops más grandes suelen costar más, aunque hay muchas excepciones con precios altos en tamaños medianos (13-15").

Peso (Weight):
Tendencia ligeramente positiva, pero no muy fuerte.
Laptops más pesadas tienden a tener mayor precio, lo cual puede reflejar mejores materiales o hardware más robusto.

Frecuencia del CPU (CPU_freq):
Relación positiva visible: procesadores de mayor frecuencia suelen encontrarse en laptops más caras.

Resolución de Pantalla (ScreenW / ScreenH):
También muestran relaciones positivas: resoluciones más altas (Full HD, Retina, etc.) suelen estar asociadas a equipos premium.

```python

# Variables a analizar respecto al precio
features = ['Ram', 'Inches', 'Weight', 'CPU_freq', 'ScreenW', 'ScreenH']

# Crear gráfico grid
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

for i, feature in enumerate(features):
    sns.scatterplot(data=laptop, x=feature, y='Price_euros', ax=axes[i], color='mediumslateblue')
    axes[i].set_title(f'Precio vs {feature}', fontsize=12)

plt.suptitle('Relación entre Precio y Variables Numéricas', fontsize=16)
plt.tight_layout()
plt.show()
```
